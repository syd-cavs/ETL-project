```python
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV 1 into Data Frame and Clean up the Data

```python
csv_file_1 = "Resources/games_data.csv"
games_df = pd.read_csv(csv_file_1)
games_df.head()
```

```python
cleaned_games_df = games_df[["game", "score", "gamers", "comp_perc", "rating", "min_comp_time", "max_comp_time"]]
cleaned_games_df
```

```python
renamed_games_df = cleaned_games_df.rename(columns={"game": "Game",
       "score": "Score", "gamers": "Gamers",    
       "comp_perc": "Completion_Percentage",
       "rating": "Rating",
       "min_comp_time": "Minimum_Completion_Time",
       "max_comp_time": "Maximum_Completion_Time"})
renamed_games_df
```

# Store CSV 2 into Data Frame and Clean up the Data

```python
csv_file_2 = "Resources/PS4_GamesSales.csv"
sales_df = pd.read_csv(csv_file_2, header=0, encoding='unicode_escape')
sales_df.head()
```

```python
cleaned_sales_df = sales_df[["Game", "Year", "Genre", "Publisher", "North America", "Global"]]
renamed_sales_df = cleaned_sales_df.rename(columns={"North America": "North_America"})
renamed_sales_df
```

# Combine the data in Pandas to see if how many videogames appear in each data set

```python
combined_df = pd.merge(cleaned_sales_df, renamed_games_df, how="inner")
combined_df
```

```python
combined_df.to_csv("cleaned_combined_data.csv", index = False)
```

# Extract Original Tables from Cleaned Combined Data

```python
new_games_df = combined_df[['Game', 'Score', 'Gamers', 
                            'Completion Percentage', 'Rating', 'Minimum Completion Time',
                            'Maximum Completion Time']]
new_games_df
```

```python
new_games_df.to_csv("cleaned_games_data.csv", index = False)
```

```python
new_sales_df = combined_df[["Game", "Year", "Genre", "Publisher", "North America", "Global"]]
new_sales_df
```

```python
new_sales_df.to_csv("cleaned_sales_data.csv", index = False)
```
