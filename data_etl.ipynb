```python
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV 1 into Data Frame and Clean up the Data

```python
csv_file_1 = "Resources/games_data.csv"
games_df = pd.read_csv(csv_file_1)
games_df.head()
```

```python
cleaned_games_df = games_df[["game", "score", "gamers", "comp_perc", "rating", "min_comp_time", "max_comp_time"]]
cleaned_games_df
```

```python
renamed_games_df = cleaned_games_df.rename(columns={   
       "comp_perc": "completion_percentage",
       "min_comp_time": "minimum_completion_time",
       "max_comp_time": "maximum_completion_time"})
renamed_games_df
```

# Store CSV 2 into Data Frame and Clean up the Data

```python
csv_file_2 = "Resources/PS4_GamesSales.csv"
sales_df = pd.read_csv(csv_file_2, header=0, encoding='unicode_escape')
sales_df.head()
```

```python
cleaned_sales_df = sales_df[["Game", "Year", "Genre", "Publisher", "North America", "Global"]]
renamed_sales_df = cleaned_sales_df.rename(columns={
    "Game": "game", "Year": "year", "Genre": "genre",
    "Publisher": "publisher", 
    "North America": "north_america", "Global": "global"})
renamed_sales_df
```

# Combine the data in Pandas to see if how many videogames appear in each data set

```python
combined_df = pd.merge(renamed_sales_df, renamed_games_df, how="inner")
new_combined_df = combined_df.drop(index=188)
new_combined_df
```

```python
new_combined_df.to_csv("cleaned_combined_data.csv", index = False)
```

# Extract Original Tables from Cleaned Combined Data

```python
new_games_df = new_combined_df[['game', 'score', 'gamers', 
                            'completion_percentage', 'rating', 'minimum_completion_time',
                            'maximum_completion_time']]
new_games_df
```

```python
new_games_df.to_csv("cleaned_games_data.csv", index = False)
```

```python
new_sales_df = new_combined_df[["game", "year", "genre", "Publisher", "North_America", "Global"]]
new_sales_df
```

```python
new_sales_df.to_csv("cleaned_sales_data.csv", index = False)
```

# Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/videogames"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_games_df.to_sql(name='ps4games', con=engine, if_exists='append', index=False)
```

```python
new_sales_df.to_sql(name='sales_data', con=engine, if_exists='append', index=False)
```
